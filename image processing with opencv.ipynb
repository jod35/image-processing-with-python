```python
import cv2
from skimage import color,io
```

```python
img1=cv2.imread('hsv.jpg')
img2=cv2.imread('jona.jpg') 
hsv_image=color.rgb2hsv(img2) #convert to hsv color space
```

```python
alpha=0.30
beta=0.70

#blended image
final_image=cv2.addWeighted(img1,alpha,img2,beta,0.0) #calculates the weighted sum of the image arrays


cv2.imwrite("jojo.jpg",final_image)

io.imshow(final_image)
```
